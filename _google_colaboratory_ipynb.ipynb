```python
import pandas as pd # Импорт библиотеки pandas для работы с данными
import matplotlib.pyplot as plt # Импорт библиотеки для визуализации данных

# Загрузка данных из CSV файла
def load_data(file_path):
    try:
        data = pd.read_csv(file_path) # Чтение данных из файла
        print("Данные успешно загружены.") # Сообщение об успешной загрузке
        return data  # Возвращение загруженных данных в виде DataFrame
    except FileNotFoundError:# Обработка ошибки, если файл не найден
        print("Файл не найден. Проверьте путь к файлу.")  # Обработка ошибки, если файл не найден
        return None  # Возвращение None при ошибке

# Обработка данных: вычисление среднего значения
def process_data(data): # Вычисление среднего значения по столбцу 'score'
    mean_score = data['score'].mean()# Вычисление среднего значения по столбцу 'score'
    print(f"Средний балл: {mean_score:.2f}") # Вывод среднего значения с двумя знаками после запятой
    return mean_score # Возвращение среднего значения

# Визуализация данных
def visualize_data(data):
    plt.figure(figsize=(10, 6)) # Установка размера графика
    plt.bar(data['name'], data['score'], color='skyblue')# Создание столбчатой диаграммы
    plt.xlabel('Имя') # Подпись оси X
    plt.ylabel('Баллы')# Подпись оси Y
    plt.title('Баллы студентов')# Заголовок графика
    plt.axhline(y=data['score'].mean(), color='r', linestyle='--', label='Средний балл') # Добавление линии, показывающей средний балл
    plt.legend() # Показ легенды графика
    plt.show() # Отображение графика

# Основная функция
def main():
    file_path = 'data.csv'  # Путь к файлу с данными
    data = load_data(file_path) # Загрузка данных из файла

    if data is not None: # Проверка, были ли успешно загружены данные
        process_data(data)  # Обработка данных
        visualize_data(data)# Визуализация данных

if __name__ == "__main__":
    main()  # Запуск основной функции, если этот файл является основным модулем
```

## PyDrive

Ниже вы найдете примеры использования PyDrive для аутентификации и загрузки/скачивания файлов. Больше примеров можно найти в <a href="https://pythonhosted.org/PyDrive/">документации PyDrive</a>.

```python
from pydrive.auth import GoogleAuth # Импорт класса GoogleAuth для аутентификации пользователя в Google Drive с использованием PyDrive
from pydrive.drive import GoogleDrive # Импорт класса GoogleDrive для работы с файлами в Google Drive через PyDrive
from google.colab import auth # Импорт модуля для аутентификации в Google Colab, позволяющего запросить доступ к аккаунту Google
from oauth2client.client import GoogleCredentials # Импорт класса GoogleCredentials для управления учетными данными OAuth 2.0 в Python
```

Выполните аутентификацию и создайте клиент PyDrive.


```python
auth.authenticate_user()# Запрашивает аутентификацию пользователя, предоставляя доступ к его Google аккаунту через Google Colab
gauth = GoogleAuth() # Создает объект GoogleAuth для управления аутентификацией и авторизацией в Google Drive
gauth.credentials = GoogleCredentials.get_application_default()# Получает учетные данные OAuth 2.0 по умолчанию для аутентификации приложения
drive = GoogleDrive(gauth)  # Создает объект GoogleDrive, используя ранее аутентифицированный объект GoogleAuth для работы с файлами в Google Drive
```

Создайте и загрузите текстовый файл.


```python
uploaded = drive.CreateFile({'title': 'Sample upload.txt'}) # Создает новый файл в Google Drive с заданным заголовком 'Sample upload.txt'
uploaded.SetContentString('Sample upload file content')# Устанавливает содержимое файла как строку 'Sample upload file content'
uploaded.Upload()# Загружает файл на Google Drive
print('Uploaded file with ID {}'.format(uploaded.get('id')))# Выводит сообщение с идентификатором загруженного файла на Google Drive
```

Загрузите файл по идентификатору и распечатайте его содержимое.


```python
downloaded = drive.CreateFile({'id': uploaded.get('id')}) # Создает объект файла в Google Drive с использованием ID загруженного файла 'uploaded'
print('Downloaded content "{}"'.format(downloaded.GetContentString()))# Получает содержимое файла и выводит его в формате строки
```

## Drive REST API

Чтобы использовать Drive API, сначала нужно выполнить аутентификацию и создать клиент API.


```python
from google.colab import auth # Импорт модуля auth из Google Colab для аутентификации пользователя
auth.authenticate_user() # Запрашивает аутентификацию пользователя, позволяя ему предоставить доступ к аккаунту Google
from googleapiclient.discovery import build # Импорт функции build для создания сервиса API Google
drive_service = build('drive', 'v3') # Создает объект сервиса API Google Drive для работы с версиями API v3
```

С этим клиентом мы можем использовать любые функции из <a href="https://developers.google.com/drive/v3/reference/">документации по Google Drive API</a>. Примеры представлены ниже.


### Создание файла на Диске на основе данных из Python

Прежде всего, создайте локальный файл для загрузки.

```python
with open('/tmp/to_upload.txt', 'w') as f:# Открывает (или создаёт) файл 'to_upload.txt' в папке '/tmp' для записи ('w')
  f.write('my sample file') # Записывает строку 'my sample file' в файл

print('/tmp/to_upload.txt contains:') # Выводит сообщение о содержимом файла
!cat /tmp/to_upload.txt # Выполняет команду 'cat' в терминале для отображения содержимого файла 'to_upload.txt'
```

Загрузите его, используя метод <a href="https://developers.google.com/drive/v3/reference/files/create"><code>files.create</code></a>. Более подробная информация о загрузке файлов доступна в <a href="https://developers.google.com/drive/v3/web/manage-uploads">документации для разработчиков</a>.

```python
from googleapiclient.http import MediaFileUpload # Импорт класса MediaFileUpload для загрузки медиафайлов в Google Drive

file_metadata = {# Создает метаданные для файла, который будет загружен
  'name': 'Sample file', # Указывает имя файла в Google Drive
  'mimeType': 'text/plain' # Указывает MIME-тип файла как текстовый файл
}
media = MediaFileUpload('/tmp/to_upload.txt', # Создает объект MediaFileUpload для файла, который будет загружен
                        mimetype='text/plain',# Указывает MIME-тип файла
                        resumable=True) # Настраивает возможность продолжения загрузки в случае её прерывания
created = drive_service.files().create(body=file_metadata, # Создает новый файл в Google Drive с заданными метаданными
                                       media_body=media,  # Указывает тело запроса как загружаемый медиафайл
                                       fields='id').execute()# Выполняет запрос и получает результат, только с полем 'id'
print('File ID: {}'.format(created.get('id'))) # Выводит идентификатор загруженного файла для ссылки на него в будущем
```

После выполнения кода в ячейке выше на странице <a href="https://drive.google.com/">https://drive.google.com/</a> появится новый файл под названием Sample file.
