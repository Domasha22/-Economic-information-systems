```python
import yfinance as yf

import matplotlib.pyplot as plt


# Получение данных для акций Apple

data = yf.download('WBD', '2025-01-01', '2025-2-25')


# Визуализация скорректированной цены закрытия

data['Close'].plot(c='#270847')

plt.title('Скорректированная цена закрытия Warner Bros. Discovery, Inc. (WBD)')

plt.grid(color='#b591b5', linestyle=':') # изменение цвета сетки

plt.xlabel('Дата')

plt.ylabel('Цена')

# Изменение цвета фона графика
plt.gca().set_facecolor('#f0f0f0')  # Измените '#f0f0f0' на нужный вам цвет

plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Создание списка тикеров акций, для которых мы хотим получить данные.
# Здесь указаны примеры тикеров (Apple, Microsoft, Google), замените их на интересующие вас.
tickers_list = ['AAPL', 'MSFT', 'GOOGL']  # пример тикеров, замените на свои

# Создание пустого DataFrame, в который будут помещены данные о ценах закрытия акций.
data = pd.DataFrame()

# Цикл для загрузки данных по каждому тикеру из списка.
for ticker in tickers_list:
    # Используем yfinance для загрузки данных о ценах закрытия акций за период с 2020-01-01 по 2022-12-31.
    # Данные о ценах закрытия для каждого тикера добавляются в DataFrame 'data'.
    data[ticker] = yf.download(ticker, start='2020-01-01', end='2022-12-31')['Close']

# Визуализация полученных данных.
# Задаем цвета для графиков каждой акции.
colors = ['#008080', '#31121a', '#896d6c']  # Здесь задаются цвета для графиков

# Строим график цен закрытия акций.
data.plot(figsize=(10, 7), color=colors)

# Настройка внешнего вида графика:
#  - Добавляем сетку для облегчения чтения данных.
plt.grid(color='#769293', linestyle=':') # изменение цвета сетки
#  - Устанавливаем цвет фона графика.
plt.gca().set_facecolor('#c6d7d8')  # Здесь задется цвет для фона графика
#  - Добавляем заголовок графика.
plt.title('Скорректированная цена закрытия')
#  - Подписываем ось X.
plt.xlabel('Дата')
#  - Подписываем ось Y.
plt.ylabel('Цена')
#  - Отображаем легенду (названия акций).
plt.legend()
#  - Отображаем сам график.
plt.show()
```
